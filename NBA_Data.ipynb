# Scrapping Web For NBA Data

- In this notebook, I will be taking information from their website and making models to predict results for the upcoming NBA season(21-22).

- NBA Site: https://www.basketball-reference.com/leagues/NBA_2021_totals.html


# Importing

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Requesting contents


```python
url = 'https://www.basketball-reference.com/leagues/NBA_2021_totals.html'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

# Getting contenet from webpage


# Finding the Full Table

```python
table = soup.find_all(class_="full_table")
table
```

# Finding the column names within the table

```python
head = soup.find(class_="thead")

column_names_raw=[head.text for item in head][0]

column_names_raw
```

# Cleaning column names

```python
column_names_clean = column_names_raw.replace("\n",",").split(",")[2:-1]

column_names_clean
```

# Retreving all player stats


```python
players = []

for i in range(len(table)):
    
    player_ = []
    
    for td in table[i].find_all("td"):
        player_.append(td.text)
        
    players.append(player_)
    
df = pd.DataFrame(players, columns = column_names_clean).set_index("Player")

df.index = df.index.str.replace('*', '')
```

```python
df
```

# Breaking down to CSV file

```python
df.to_csv ('2021_nba_data_total.csv',header=True)
```
